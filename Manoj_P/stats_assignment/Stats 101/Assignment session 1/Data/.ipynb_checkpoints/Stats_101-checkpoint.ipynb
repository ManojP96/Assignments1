# Data Cleaning

```python
duplicated_videos_pb=(video['video_id']+video['publish_time']).value_counts() #checking for duplicate records
```

```python
len(duplicated_videos_pb[duplicated_videos_pb>1]) 
```

Same video is repeated multiple times

```python
duplicated_videos_tr=(video['video_id']+video['publish_time']+video['trending_date']).value_counts()
len(duplicated_videos_tr[duplicated_videos_tr>1]) #checking if treding date for repeated videos are same  
```

We have same video multiple times but with different trending date which indicates same video 
might be trended multiple times

So it is best to remove the video which have trended multiple times since we are analysing trending pattren

```python
video.sort_values(by=['publish_time'],ascending=True,inplace=True)#sorting based on publish time
```

```python
repeated_videos=video[video.duplicated(subset=['video_id','publish_time'])]#storing repeated videos data
```

```python
video.drop_duplicates(subset=['video_id','publish_time'],keep='first',inplace=True)# droping the repeated videos
```

```python
video.shape
```

# Cleaning Date time format

```python
import datetime
from datetime import datetime
```

```python
video['trending_date']
```

```python
video['trending_date']=video['trending_date'].map(lambda x:datetime.strptime(x, '%y.%d.%m'))
```

```python
video['publish_time']=video['publish_time'].map(lambda x: x.split('T')[0]).map(lambda x:datetime.strptime(x, '%Y-%m-%d'))
```

```python
video['publish_time']=pd.to_datetime(video['publish_time'])
```

```python
trending_duration=(video['trending_date']-video['publish_time']).dt.days
```

# How long does it take for the video to get trending

checking distribution of views

```python
sns.kdeplot(trending_duration)
```

```python
sns.kdeplot(trending_duration,clip=(0,100))
```

```python
print('avg time in days for video to get trending is {}'.format(np.ceil(trending_duration.mean())))
print('median of trending duration is',(trending_duration).median())
```

```python
len(trending_duration)#total videos uploaded
```

```python
len(trending_duration[trending_duration>3]) #to videos which started trending more than 3 days of publishing
```

So on an avrage it takes 1 to 3 days for video to get trending but very few videos took more then 3 days to get trending

# How is this duration changes across the two years

```python
video['publish_time'].dt.year.value_counts()
```

very few videos are uploaded till 2016 and we can see drastic change in 2017 and 2018

```python
past_two_year=video['publish_time'].dt.year[video['publish_time'].dt.year>2015] #filtering videos published after 2016
```

# Analysing views, comment, dislikes and likes

```python
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
sns.scatterplot(y=video['views'],x=video['comment_count'])
plt.subplot(2,2,2)
sns.scatterplot(y=video['views'],x=video['likes'])
plt.subplot(2,2,4)
sns.heatmap(video[['views','likes','dislikes','comment_count']].corr(),cmap='Blues',annot=True,
           mask=np.triu(video[['views','likes','dislikes','comment_count']].corr()))
plt.subplot(2,2,3)
sns.scatterplot(y=video['views'],x=video['dislikes'])
```

Views increase with more likes and also with more comments
we can see views and likes have strong correlation of 0.78 and views and comment count are moderately correlated at 0.68
and also we can observe "there is no negetive effect on dislikes on views"

# Checking number of videos uploaded in each category

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
video.groupby(by=['category_id'])['views'].mean().sort_values(ascending=False).plot(kind='bar')
plt.title('Avg_views_per_Category')
plt.ylabel('Avg_views')
plt.xticks(rotation=360);
plt.subplot(1,2,2)
video['category_id'].value_counts().plot(kind='bar')
plt.title('Video per Category')
plt.xlabel('cat_id')
plt.ylabel('Count_of_video')
plt.xticks(rotation=360);
```

most viewed category is 10 and most uploaded category is 24

Very less videos are uploaded in category 43 but its has second higest views
