```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from datetime import datetime
```

```python
df= pd.read_csv('Data/Videos2.csv')
```

```python
df.head()
```

```python
date_conversion=lambda x:datetime.strptime(x, '%y.%d.%m')
```

```python
df['publish_time']=df['publish_time'].map(lambda x: x.split('T')[0]).map(lambda x:datetime.strptime(x, '%Y-%m-%d'))
df['publish_time']=pd.to_datetime(df['publish_time'])
df["publish_time"].dtype
```

```python
df.info()
```

```python
df['publish_time']
```

```python
df.info()
```

```python
df['publish_date']=df['publish_time']
```

```python
df.info()
```

```python
#Function to calculate whether the video published on weekday or weekend
def weekt(day):
    if day=="Sunday" or day =="Saturday":
        return "Weekend"
    else:
        return "Weekday"
    
```

```python
#Creating a column week_time
df['week_time']=np.vectorize(weekt)(df['publish_date'])
```

```python
df['trending_date']
```

```python
'''making three columns for trening date - year as 'y', date as 'd', month as'm' as trending date format
is not supported by datetime'''
df[['y','d','m']]=df['trending_date'].str.split('.',expand=True)
```

```python
df.head()
```

```python
#checking null values
df.isnull().sum()
```

```python
df['description'].fillna(" No Description ",inplace=True)
```

```python
#checking null values
df.isnull().sum()
```

```python
#Creating a string trending date which can be further used to calculate trending date using to_datetime
df['trending_date1']='20'+df['y']+'-'+df['d']+'-'+df['m']
```

```python
df['publish_date']
```

```python
df['trending_date1']
```

```python
#Formatting trending_date1 using to_datetime and using format as YYYY-MM-DD
df['trending_date1']= pd.to_datetime(df['trending_date1'], format='%Y-%m-%d')
```

```python
df['trending_date1']
```

```python
df.info()
```

```python
#Calculating how many days it tooks for a video to trend
df['days_to_trend']=(df['trending_date1']-df['publish_date']).dt.days
```
