```python
import yfinance as yf
import pandas as pd
```

```python
stock_names=['BHARTIARTL.NS',"SBIN.NS","DRREDDY.NS","GSPL.NS",
             "JSWSTEEL.NS","JSWENERGY.BO","RELIANCE.NS","YESBANK.NS",
             "TATAMOTORS.NS","LAURUSLABS.NS","NTPC.NS","SAIL.BO"]
```

```python
final_stocks=pd.DataFrame()
for stock in stock_names:
    stocks = yf.download(stock, interval='1d', start="2015-01-01", end="2021-12-31")
    stocks['Stock_name']=stock
    final_stocks=pd.concat([final_stocks,stocks])
```

```python
final_stocks.info()
```

```python
final_stocks.reset_index(inplace=True)
```

```python
final_stocks.columns
```

```python
final_stocks.drop(['High','Low','Volume','Adj Close'],axis=1,inplace=True)
```

```python
final_stocks.head()
```

```python
final_stocks['Profit%'] = (final_stocks['Close']-final_stocks['Open'])/final_stocks['Open']*100
final_stocks
```

```python
final_stocks = final_stocks[final_stocks['Profit%']>5.8]
final_stocks
```

```python
final_stocks.sort_values(by=['Date','Profit%'], ascending = [True,False],inplace = True)
final_stocks
```

```python
final_stocks.drop_duplicates(subset=['Date'],inplace=True)
final_stocks
```

```python
capital=100000
transaction= pd.DataFrame(['Date','Name of the stocks','Transaction','Event','Total Units','Capital'])
```

```python
transaction = pd.DataFrame(columns=['Date', 'Stock_name', 'Transaction_price', 'Event', 'Total_unit', 'Net_worth'])
capital=100000

for inv in range(final_stocks.shape[0]):
    Stock_name = final_stocks.iloc[inv]['Stock_name']
    total_units = capital // final_stocks.iloc[inv]['Open']
    transaction_price = total_units *final_stocks.iloc[inv]['Close']*0.05
    event='Buy'
    transaction = transaction.append({'Date': final_stocks.iloc[inv]['Date'], 
                                        'Stock_name': Stock_name, 
                                        'Transaction_price': 0, 
                                        'Event': event, 
                                        'Total_unit': total_units, 
                                        'Net_worth': capital}, ignore_index=True)
    profit = total_units * (final_stocks.iloc[inv]['Close']-final_stocks.iloc[inv]['Open']) - (total_units * final_stocks.iloc[inv]['Close']) * 0.05
    capital=capital+profit
    event='Sell'
    transaction = transaction.append({'Date': final_stocks.iloc[inv]['Date'], 
                                        'Stock_name': Stock_name, 
                                        'Transaction_price': transaction_price, 
                                        'Event': event, 
                                        'Total_unit': total_units, 
                                        'Net_worth': capital}, ignore_index=True)
```
